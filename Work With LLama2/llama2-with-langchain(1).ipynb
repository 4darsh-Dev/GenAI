```python
from langchain.llms import HuggingFacePipeline
```

```python
from transformers import AutoTokenizer
```

```python
import torch
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model="meta-llama/Llama-2-7b-chat-hf"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
import transformers
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    
)
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature': 0.7})
```

```python
prompt = "What is the good name for company that make colourful socks"
```

```python
from langchain.prompts import PromptTemplate
```

```python
from langchain.chains import LLMChain
```

```python
prompt_template1=PromptTemplate(input_variables=["cuisine"],
                               template="I want to open a restaurant for {cuisine} food. Suggest a fancy name for this.")
```

```python
input_prompt=prompt_template1.format(cuisine="indian")
```

```python
print(input_prompt)
```

```python
prompt_template2=PromptTemplate(input_variables=["book_name"],
                               template="Provide me a summary of {book_name}")
```

```python
chain=LLMChain(llm=llm, prompt=prompt_template2, verbose=True)
response = chain.run("Harry Potter")
```
