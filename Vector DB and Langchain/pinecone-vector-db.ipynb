```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Load model and tokenizer
model_name = "gpt2"  # You can replace this with any other model from Hugging Face
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Create a pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=100
)

# Create a LangChain wrapper
llm = HuggingFacePipeline(pipeline=pipe)

# Use the model
response = llm("What is artificial intelligence?")
print(response)
```

### downloading the pdf and storing 

```python
import os
import requests

# Create a 'pdfs' folder if it doesn't exist
if not os.path.exists('pdfs'):
    os.makedirs('pdfs')

# URL of the PDF you want to download
pdf_url = "https://proceedings.neurips.cc/paper_files/paper/2017/file/3f5ee243547dee91fbd053c1c4a845aa-Paper.pdf"

# Get the filename from the URL
filename = pdf_url.split("/")[-1]

# Full path where the PDF will be saved
save_path = os.path.join('pdfs', filename)

# Download the PDF
response = requests.get(pdf_url)

# Check if the request was successful
if response.status_code == 200:
    # Write the content to a file
    with open(save_path, 'wb') as file:
        file.write(response.content)
    print(f"PDF downloaded and saved to {save_path}")
else:
    print(f"Failed to download PDF. Status code: {response.status_code}")
```

```python
loader = PyPDFDirectoryLoader("pdfs")
```

```python
data = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 20
)
```

```python
text_chunks = text_splitter.split_documents(data)
```

```python
# Prepare the data for upsert
batch_size = 100
for i in range(0, len(text_chunks), batch_size):
    batch = text_chunks[i:i+batch_size]
    
    vectors_to_upsert = []
    for j, doc in enumerate(batch):
        text = doc.page_content
        embed = embeddings.embed_query(text)
        
        vector = {
            "id": f"id_{i+j}",
            "values": embed,
            "metadata": {"text": text}
        }
        vectors_to_upsert.append(vector)
    
    # Upsert to Pinecone
    index.upsert(vectors=vectors_to_upsert, namespace="ns1")
```
